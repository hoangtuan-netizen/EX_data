 Pixel Transforms (Biến đổi điểm ảnh)
Giải thích: Thao tác trên từng pixel riêng lẻ. Công thức tổng quát:
O = a * I + b
Trong đó:

a (gain): Điều chỉnh độ tương phản.

b (bias): Điều chỉnh độ sáng.

Ví dụ: Tăng độ sáng và độ tương phản của ảnh.

```python
import cv2
import numpy as np

# Đọc ảnh
img = cv2.imread('Meo.jpg')
# Tham số
a = 1.5  # Tăng độ tương phản
b = 30   # Tăng độ sáng
# Áp dụng công thức
adjusted = np.clip(a * img + b, 0, 255).astype(np.uint8)
# Lưu ảnh
cv2.imwrite('adjusted.jpg', adjusted)
```

Color Transforms (Biến đổi màu sắc)
Giải thích: Chuyển ảnh màu sang ảnh xám dựa trên trọng số RGB.
Công thức:
Gray = 0.299 * R + 0.587 * G + 0.114 * B

Ví dụ: Chuyển ảnh RGB sang grayscale.

```python
import cv2

# Đọc ảnh màu
img = cv2.imread('input.jpg')
# Chuyển sang grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Lưu ảnh
cv2.imwrite('gray.jpg', gray)
```

 Histogram Equalization (Cân bằng histogram)
Giải thích: Cải thiện độ tương phản bằng cách phân phối lại giá trị pixel.
Ví dụ: Cân bằng histogram cho ảnh xám.

```python
import cv2

# Đọc ảnh xám
gray = cv2.imread('gray.jpg', 0)
# Cân bằng histogram
equalized = cv2.equalizeHist(gray)
# Lưu ảnh
cv2.imwrite('equalized.jpg', equalized)
```

Thresholding (Phân ngưỡng)
Giải thích: Chuyển ảnh thành nhị phân dựa trên ngưỡng T.
Ví dụ: Phân ngưỡng cố định với T = 127

```python
import cv2

# Đọc ảnh xám
gray = cv2.imread('gray.jpg', 0)
# Áp dụng phân ngưỡng
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# Lưu ảnh
cv2.imwrite('binary.jpg', binary)
```

Contrast Stretching (Kéo giãn độ tương phản)
Giải thích: Ánh xạ giá trị pixel từ [min, max] sang [0, 255].
Công thức:
O = (I - min) * (255 / (max - min))

Ví dụ: Kéo giãn độ tương phản cho ảnh tối.

```python
import cv2
import numpy as np

# Đọc ảnh xám
gray = cv2.imread('dark_image.jpg', 0)
# Tính min và max
min_val = np.min(gray)
max_val = np.max(gray)
# Áp dụng công thức
stretched = ((gray - min_val) / (max_val - min_val) * 255).astype(np.uint8)
# Lưu ảnh
cv2.imwrite('stretched.jpg', stretched)
```

 Image Arithmetic (Số học ảnh)
Giải thích: Thực hiện phép toán số học trên ảnh (cộng, trừ, nhân, chia).
Ví dụ: Trộn hai ảnh với tỉ lệ alpha.

```python
import cv2

# Đọc hai ảnh
img1 = cv2.imread('img1.jpg')
img2 = cv2.imread('img2.jpg')
# Trộn ảnh với alpha = 0.5
blended = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
# Lưu ảnh
cv2.imwrite('blended.jpg', blended)
```

Khi chuyển từ ảnh màu sang ảnh xám ta có thể dùng công thức: x = r * 0.299 + g * 0.587 +
b * 0.114

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc ảnh màu (OpenCV đọc ảnh theo dạng BGR)
image = cv2.imread('Lenna.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Chuyển sang ảnh xám theo công thức chuẩn
gray = 0.299 * image_rgb[:, :, 0] + 0.587 * image_rgb[:, :, 1] + 0.114 * image_rgb[:, :, 2]
gray = gray.astype(np.uint8)  # Đảm bảo định dạng ảnh uint8

# Hiển thị ảnh gốc và ảnh xám
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Ảnh màu')

plt.subplot(1, 2, 2)
plt.imshow(gray, cmap='gray')
plt.title('Ảnh xám')

plt.show()
```
